```python
import numpy as np
import arviz as az
import pymc3 as pm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.rcParams['figure.figsize'] = [15, 2]
temperatures = np.array([10, 7, 5, 6, 0, -2, -5, -1, 1, -5, -11, -12, -15, -6, -4, 1, 0, -2, -2, -5, -6, 0, 1, 5, 5, 6])
emissions = np.array([25, 50, 100, 50, 125, 150, 200, 175, 100, 150, 375, 400, 450, 325, 275, 200, 225, 250, 275, 300, 325, 200, 175, 150, 150, 150])
plt.scatter(np.arange(len(temperatures)), temperatures)
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
plt.scatter(np.arange(len(emissions)), emissions)
plt.xlabel("Day")
plt.ylabel("Emissions")
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [12, 6]

ls_coef_ = np.cov(temperatures, emissions)[0,1]/np.var(temperatures)
ls_intercept = emissions.mean() - ls_coef_*temperatures.mean()
print(f"{ls_coef_:.3f}X + {ls_intercept:.3f}")
plt.plot(temperatures, ls_coef_*temperatures + ls_intercept, label ="Least-squares prediction", color='k')
plt.scatter(temperatures, emissions)
plt.xlabel("Temperature")
plt.ylabel("Emissions")
plt.show()
```

```python
with pm.Model() as model:
    b= pm.Uniform("b", lower=-100, upper=100)    
    m_e= pm.Deterministic("m_e", b*15-b*temperatures)
    sigma_e= pm.Uniform("sigma_e", lower=0.1, upper=100)
    emission = pm.Normal("emission", mu=m_e, sigma=sigma_e, observed=emissions)
    idata= pm.sample(2000, tune=2500)
```

```python
az.plot_posterior(idata, var_names=["sigma_e", "b"], figsize=[12, 8])
```

```python
with pm.Model() as model:
    b_1 = pm.Uniform("b_1", lower=-100, upper=100)
    b_2 = pm.Uniform("b_2", lower=-100, upper=100)
    tau = pm.DiscreteUniform("tau", lower=0, upper=len(temperatures) -1)
    idx= np.arange(len(temperatures)) 
    b = pm.math.switch(tau > idx, b_1, b_2)
    m_e= pm.Deterministic("m_e", b*15-b*temperatures)

    sigma_e= pm.Uniform("sigma_e", lower=0.1, upper=100)
    emission = pm.Normal("emission", mu=m_e, sigma=sigma_e, observed=emissions)
    idata= pm.sample(2000, tune=2500)
```

```python
az.plot_posterior(idata, var_names=["sigma_e", "b_1", "b_2", "tau"], figsize=[12, 8])
```

Scenario:

Every month you tank your new car and drive it for the whole month. At the end of the month, you note down the kilometers travelled and liters tanked. After some time you feel that the fuel consumption has increased - you try to determine the moment when the fuel consumption increased - maybe it will be covered in the warrany.


```python
from numpy.random import normal as nr
fuel_cons_1 = 6.5
fuel_cons_2 = 8

b1 = 0
b2 = 0

days = 50
change = 28

fuel_tanked = np.random.normal(loc=75, scale=5, size=(days,))

km_travelled = np.concatenate((
    (fuel_cons_1 + nr(scale=0.3, size = (fuel_tanked[:change].shape)))* fuel_tanked[:change] + b1 ,
    (fuel_cons_2 + nr(scale=0.3, size = (fuel_tanked[change:].shape)))* fuel_tanked[change:] + b2
))

plt.plot(fuel_tanked, 'o')
plt.xlabel("Months")
plt.ylabel("Liters Tanked")
plt.show()

plt.plot(km_travelled, 'o')
plt.xlabel("Months")
plt.ylabel("km Travelled")
plt.show()

plt.scatter(fuel_tanked, km_travelled)
ls_coef_ = np.cov(fuel_tanked, km_travelled)[0,1]/np.var(fuel_tanked)
ls_intercept = km_travelled.mean() - ls_coef_*fuel_tanked.mean()
print(f"{ls_coef_:.3f}X + {ls_intercept:.3f}")
plt.plot(fuel_tanked, ls_coef_*fuel_tanked + ls_intercept, label ="Least-squares prediction", color='k')
plt.xlabel("Fuel Tanked")
plt.ylabel("km Travelled")
plt.show()
```

```python
with pm.Model() as model:
    a= pm.Uniform("a", lower=-10, upper=10)    
    m_cons= pm.Deterministic("m_cons", a*fuel_tanked)
    sigma_cons= pm.Uniform("sigma_cons", lower=0.01, upper=10000)
    cons = pm.Normal("fuel_consumption", mu=m_cons, sigma=sigma_cons, observed=km_travelled)
    idata= pm.sample(2000, tune=2500)
```

```python
az.plot_posterior(idata, var_names=["sigma_cons", "a"], figsize=[12, 8])
```

```python
with pm.Model() as model:
    a_1 = pm.Uniform("a_1", lower=-10, upper=10)
    a_2 = pm.Uniform("a_2", lower=-10, upper=10)
    tau = pm.DiscreteUniform("tau", lower=0, upper=len(fuel_tanked)-1)
    idx= np.arange(len(fuel_tanked)) 
    a = pm.math.switch(tau > idx, a_1, a_2)
    m_cons= pm.Deterministic("m_cons", a*fuel_tanked)

    sigma_cons= pm.Uniform("sigma_cons", lower=0.1, upper=10000)
    cons = pm.Normal("fuel_consumption", mu=m_cons, sigma=sigma_cons, observed=km_travelled)
    idata= pm.sample(2000, tune=2500)
```

```python
az.plot_posterior(idata, var_names=["sigma_cons", "a_1", "a_2", "tau"], figsize=[12, 8])
```

In this excercise, we observe the data, that we assume follows a linear function. In one point in time, we suspect that something has changed - the coefficient of linear function has increased.

First we check the linear model - we can see that the standard deviation (sigma) is very high - indicating that the prediction is poor. And in fact we can see that the line does not represent the data well, since the original data is composed of essentially 2 functions - one linear up to a switch point, then a second linear.

To approximate it better we use two functions with variable switch point. We feed that to the model and obtain correct predictions - two linear coefficient and the switch poinnt is calculated automatically.

#### Scenario:

##### Every month you tank your new car and drive it for the whole month. At the end of the month, you note down the kilometers travelled and liters tanked. After some time you feel that the fuel consumption has increased - you try to determine the moment when the fuel consumption increased - maybe it will be covered in the warrany.

Having this data I prepared a similar model with two linear coefficients and a variable switching point. The switch point was predicted accurately and the slopes of the lines matched the data I prepared. However, due to the data, the sigma was still quite high

Next, I tried to make the switch less pronounced. With the 2 coefficents more similar to one another, the sigma decreased and prediction was more accurate.

Then, I tried to increase the noise. As expected, more noise led to much poorer prediction, affecting both slopes (a_1, a_2), sigma and even the switch point

Finally, I tried to make the change more gradual. I made the values of a_1 go gradually from 6.5 to 8 over the course of 8 days. This has made a huge impact on the prediction, which it missed by 10 days, and the slopes were also very highly varied with huge error margins.
