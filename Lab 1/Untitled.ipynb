```python
import matplotlib.pyplot as plt
import numpy as np
import math
import scipy.stats

fig, ax = plt.subplots(1, 2, figsize=(20,6))

for mu, sigma in [(0, 0.3), (0, 0.5), (0, 0.7), (0, 0.9), (0, 2)]:
    x = np.arange(-5, 5, 0.01)
    pdf = scipy.stats.norm.pdf(x, mu, sigma)
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('PDF (Normal)')
    ax[0].plot(x, pdf, label=f'$\mu$: {mu}, $\sigma$: {sigma}')
    ax[0].legend()
    cdf = scipy.stats.norm.cdf(x, mu, sigma)
    ax[1].set_xlabel('x')
    ax[1].set_ylabel('CDF (Normal)')
    ax[1].plot(x, cdf, label=f'$\mu$: {mu}, $\sigma$: {sigma}')
    ax[1].legend()
```

```python
plt.figure(figsize=(10,6))
x = np.arange(0, 10, 1)
pmf_1 = scipy.stats.binom.pmf(x, 10, 0.5)
pmf_2 = scipy.stats.binom.pmf(x, 10, 0.1)
plt.xlabel('x')
plt.ylabel('PMF')
plt.plot(x, pmf_1,'-o', label=f'Binom(10, 0.5)', )
plt.plot(x, pmf_2,'--x', label=f'Binom(10, 0.1)')
plt.legend()
```

```python
plt.figure(figsize=(10,6))
x = np.arange(0, 20)
for m in [0.1, 2, 5, 10]:
    pmf = scipy.stats.poisson.pmf(x, m)
    plt.plot(x, pmf, '-o', label=f'$\mu$ = {m}')
    plt.legend()
```
